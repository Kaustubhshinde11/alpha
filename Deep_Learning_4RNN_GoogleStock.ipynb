```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("/Users/kaustubhshinde/Downloads/Google_Stock_Price_Train.csv")
```

```python
df
```

```python

df['Close'] = df['Close'].str.replace(',', '').astype(float) #replace commas and spaces 
data = df["Close"].values   # Gives you values inside the array.
data = data.reshape(-1, 1)   #It reshapes the data -1 implies take no.of.rows of your choice but take columns only 1 
```

```python

sc = MinMaxScaler()                       #Stock prices generally have high values so we scale them between 0 and 1 to ease computation
scaled_data = sc.fit_transform(data)
```

```python
x = []                      # Here first I created 2 lists in x I am going to store the current sock values till date and in y 
y = []                      # store the next predicted stock values so by running loop I appended the current stock values in x.
for i in range(60, len(scaled_data)):        
    x.append(scaled_data[i-60:i])
    y.append(scaled_data[i])

x = np.array(x)     # Here I just created an array of x and y values.
y = np.array(y)
```

```python
split = int(len(x)*0.8)                          #this is called chronological splitting in our conventional splitting the split 
x_train, x_test = x[:split], y[split:]           # doesn't preserve the order but here it preserves and we need it.
y_train, y_test = y[:split], y[split:]
```

```python
model = Sequential([
    LSTM(50, return_sequences = False, input_shape = (60, 1)),    #LSTM layer is a specific layer which is used to build the recurrent neural network
    Dense(1)                                                    # we provide 50 neurons and last 60 days prices to produce 1 dense output layer.
])
```

```python
model.compile(optimizer = "adam", loss = "mean_squared_error")
```

```python
model.fit(x_train, y_train, epochs = 10)
```

```python
loss = model.evaluate(x_test, y_test)
```

```python
predict = model.predict(x_test)
```
